```python
import numpy as np
import scipy.integrate
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from matplotlib import cm

# Problem 1
#Define ODE
def rhsfunc1(t, y, eps):
    f1 = y[1]
    f2 = (t**2 - eps)*y[0]
    return np.array([f1, f2])

# Define some constants
#n0 = 0 #defined inside the function
xp = [-4, 4]
tol = 1e-6
# Define our initial conditions
A = 1

x_evals = np.linspace(-4, 4, 81)
epsilon_start = 0# This is our initial beta value, we will change it, 
#recommended on piazza to start from epsilon = 0
eigenvalues1 = np.zeros(5)
eigenfunctions1 = np.zeros([81, 5])
eigenfunctions = np.zeros([81, 5])

# Make a loop over beta values to find more eigenvalue-eigenfunction pairs
#modes is another way to say eigenfunction
for mode in range(5):# Try to find 5 modes
    epsilon = epsilon_start
    depsilon = 1
    
    for i in range(1000):
        y0 = np.array([A, A*np.sqrt(4**2 - epsilon)])
        sol = scipy.integrate.solve_ivp(lambda x, y: rhsfunc1(x, y, epsilon), xp, y0, t_eval = x_evals)

        if (np.abs(sol.y[1, -1] + np.sqrt(4**2 - epsilon)*sol.y[0, -1])) < tol:
            #the boundary condition at phi'(x=L) should be limited to be less than here
            #phi'(L) = - sqrt(epsilon)*phi(L) -->given < tol
            #print(r'We got the eigenvalue! $\epsilon = $', epsilon)
            eigenfunctions1[:, mode] = sol.y[0, :]
            eigenvalues1[mode] = epsilon
            break

        if (-1)**(mode)*(sol.y[1, -1] + np.sqrt(4**2 - epsilon)*sol.y[0, -1]) > 0:
            #phi'(L) = - sqrt(KL^2 - epsilon)*phi(L)
            epsilon = epsilon + depsilon
            # Decrease beta if y(1)>0, because we know that y(1)>0 for beta = beta_start
        else:
            epsilon = epsilon - depsilon/2 # Increase beta by a smaller amount if y(1)<0
            depsilon = depsilon/2 # Cut dbeta in half to make we converge
    
    epsilon_start = epsilon + 0.1# increase beta once we have found one mode.
    eig_norm = np.trapz(eigenfunctions1[:, mode]**2, x = x_evals)
    eigenfunctions1[:, mode] = eigenfunctions1[:, mode]/np.sqrt(eig_norm)
    
    plt.plot(sol.t, eigenfunctions1[:, mode], linewidth=2)
    plt.plot(sol.t, 0*sol.t, 'k')

    eigenfunctions[:, mode] = eigenfunctions1[:, mode]
    eigenfunctions1[:, mode] = abs(eigenfunctions1[:, mode])
```

```python
## 3D Plot
%matplotlib notebook
#defining the axes
dx = np.linspace(-4, 4, 81)
dt = np.linspace(0, 5, 100)

#calculating the density function
density_func = (eigenfunctions[:, 1].reshape(-1, 1) * np.cos(eigenvalues1[1]*dt/2)).T

#print(np.shape(dx))
#print(np.shape(dt))
#print(np.shape(density))

fig = plt.figure()
ax = plt.axes(projection = "3d")
X, T = np.meshgrid(dx, dt)
surface = ax.plot_surface(X, T, density_func, cstride=1, rstride=1, cmap = cm.plasma)
fig.colorbar(surface, pad = 0.1)
ax.contour(X, T, density_func,   offset = -0.6, cmap = cm.plasma)
ax.set_xlabel("x")
ax.set_ylabel("t")
ax.set_zlabel(r"$\psi_2(x, t)$")
plt.title(r"Time evolution of the second mode $\psi_2(x, t)$");
```
