```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
import copy
#%matplotlib notebook
```

```python
#copy of correct one
K = 1
L = 4
xp = [-L, L]
tol = 1e-6
x = np.linspace(-L,L,20*L+1)
dx = 0.1

# construct M
M = (1/dx**2) * (np.diag(2*np.ones(20*L-1)) + np.diag(-1*np.ones(20*L-2), 1) + np.diag(-1*np.ones(20*L-2), -1))
M[0,0:2] = [2/3/dx**2,-2/3/dx**2]
M[78,77:79] = [-2/3/dx**2,2/3/dx**2]
for i in range(20*L-1) : 
    M[i,i] = M[i,i] + K*(x[i+1])**2
print(M)
# sort eigenvalues and eigenvectors 
w1, v1 = np.linalg.eig(M)
idx = w1.argsort()[::1]   
w = w1[idx]
v = v1[:,idx]

v0 = np.zeros(5)
v81 = np.zeros(5)
v_norm = []

for i in range(5):
    v0[i] = (4*v[0,i]-v[1,i])/(2*dx*np.sqrt(K*L**2-w[i])+3)
    v81[i] = (v[77,i]-4*v[78,i])/(-2*dx*np.sqrt(K*L**2-w[i])-3)
    
    #v_temp = np.insert(v[:,i], 0, v0[i])
    #v_full = np.insert(v_temp, 80, v81[i])
    
    v_temp = np.append(v[:,i], v0[i])
    v_full = np.append(v_temp, v81[i])
    
    norm = scipy.integrate.trapz(v_full*v_full,x)
    vnorm = v_full / np.sqrt(norm)
    v_norm.append(vnorm.reshape(-1,1))
```

```python
x = np.linspace(-4, 4, 20*4+1)
fig, ax = plt.subplots()
ax.plot(x, v_norm[0], linewidth=3)
ax.plot(x, v_norm[1], linewidth=3)
ax.plot(x, v_norm[2], linewidth=3)
ax.plot(x, v_norm[3], linewidth=3)
ax.plot(x, v_norm[4], linewidth=3)
ax.legend([r"$\phi_1$", r"$\phi_2$",  r"$\phi_3$",  r"$\phi_4$", r"$\phi_5$"])
```
